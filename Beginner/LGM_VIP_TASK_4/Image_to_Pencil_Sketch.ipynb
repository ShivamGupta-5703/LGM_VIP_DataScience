# **TASK - 4 --> Image to Pencil Sketch Python Project :**

Import Library -->

```python
import cv2
from google.colab.patches import cv2_imshow
```

Read Image -->

```python
image = cv2.imread('/content/image.jpeg')
cv2_imshow(image)
cv2.waitKey(0)
```

Converting Image to GreyScale -->

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
cv2.waitKey(0)
```

Inverting GreyScale Image -->

```python
inverted_image = 255 - gray_image
cv2_imshow(inverted_image)
cv2.waitKey()
```

Blur the Image by Gaussian Function In OpenCV -->

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
```

Inverting Blurred Image to convert image into Pencil Sketch -->

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
cv2_imshow(pencil_sketch)
cv2.waitKey(0)
```

Comparing Original and Pencil Sketch Image -->

```python
cv2_imshow(image)
cv2_imshow(pencil_sketch)
cv2.waitKey(0)
```
